```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates  

# File path
file_path = 'serial_20250120_113428(RU).txt'  

# Read the data without headers and assign column names
data = pd.read_csv(file_path, names=['Tempo', 'Temperatura', 'Umidade', 'CO_ppm', 'CO2_ppm'], header=None)

data['Tempo'] = data['Tempo'].str.strip()

data['Tempo'] = pd.to_datetime(data['Tempo'], format='%H:%M:%S')

# Plot the data
plt.figure(figsize=(12, 6))

# Plot temperature
plt.plot(data['Tempo'], data['Temperatura'], label='Temperatura (°C)', color='red')
plt.xlabel('Tempo')
plt.ylabel('Temperatura (°C)')
plt.title('Temperatura')
plt.legend()
plt.grid()

# Format x-axis to show minutes and add ticks every 5 minutes
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))  # Format to HH:MM
plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=2))  # Set ticks every 5 minutes
plt.xticks(rotation=30)  # Rotate for better readability

# Save the plot as a PNG
plt.savefig('RU_temp', dpi=300, bbox_inches='tight')  # Save with high resolution

# Plot humidity
plt.figure(figsize=(12, 6))
plt.plot(data['Tempo'], data['Umidade'], label='Umidade (%)', color='teal')
plt.xlabel('Tempo')
plt.ylabel('Umidade (%)')
plt.title('Umidade Relativa')
plt.legend()
plt.grid()

# Format x-axis to show minutes and add ticks every 5 minutes
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))  # Format to HH:MM
plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=2))  # Set ticks every 5 minutes
plt.xticks(rotation=30)  # Rotate for better readability

# Save the plot as a PNG
plt.savefig('RU_umi', dpi=300, bbox_inches='tight')  # Save with high resolution


# Plot CO ppm
plt.figure(figsize=(12, 6))
plt.plot(data['Tempo'], data['CO_ppm'], label='CO (ppm)', color='dimgrey')
plt.xlabel('Tempo')
plt.ylabel('CO (ppm)')
plt.title('Concentração de Monóxido de Carbono')
plt.legend()
plt.grid()

# Format x-axis to show minutes and add ticks every 5 minutes
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))  # Format to HH:MM
plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=2))  # Set ticks every 5 minutes
plt.xticks(rotation=30)  # Rotate for better readability

# Save the plot as a PNG
plt.savefig('RU_co', dpi=300, bbox_inches='tight')  # Save with high resolution




# Plot volatile compounds
plt.figure(figsize=(12, 6))

plt.plot(data['Tempo'], data['CO2_ppm'], label='CO2 (ppm)', color='royalblue')

plt.xlabel('Tempo')
plt.ylabel('CO2 (ppm)')
plt.title('Concentração de Dióxido de Carbono')
plt.legend()
plt.grid()

# Format x-axis to show minutes and add ticks every 5 minutes
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))  # Format to HH:MM
plt.gca().xaxis.set_major_locator(mdates.MinuteLocator(interval=2))  # Set ticks every 5 minutes
plt.xticks(rotation=30)  # Rotate for better readability

# Save the plot as a PNG
plt.savefig('RU_co2', dpi=300, bbox_inches='tight')  # Save with high resolution

# Show the plot
plt.show()
```
